If $B_t$ is Brownian motion, show that

$$
X_t=e^{\int_0^t \sigma_s\,dB_s-\frac12\,\int_0^t\,\sigma_s^2\,ds}
$$

is a martingale. (Hint:  Apply Itô's Lemma to $e^Y$ where $Y=\log X$.)

Deduce that

$$
e^{\frac12\,\alpha^2\,t}=\E\left[e^{\alpha\,\sqrt{t}\,Z}\right]
%\label{eq:expmartingale}
$$

where $Z \sim N(0,1)$.  Is this last equation still true if $\alpha$ is a matrix?  Why?

### Solution 1

Let 
$$
Y = \int_0^t \sigma_s\,dB_s-\frac12\,\int_0^t\,\sigma_s^2\,ds
$$

Then
$$
dY =  \sigma_t\,dB_t-\frac12\,\,\sigma_t^2\,dt
$$

Let
$$
f = e^Y
$$

Apply Ito, we have,
\begin{eqnarray}
df &=& f_t\,dt + f_Y\,dY + \frac{1}{2}\,f_{YY}\,d[Y]_t \\
&=& f\,(-\frac12\,\sigma_t^2\,dt + \sigma_t\,dB_t) + \frac12\,f\,\sigma_t^2\,dt \\
&=& f\,\sigma_t\,dB_t
\end{eqnarray}

So, it doesn't have the drift term, Q.E.D

Suppose $\sigma_x = \alpha$, from martingale property, we have

$$
\E[X_t\,|\,\mathscr{F_0}] = X_0 = 1
$$

which means,
$$
\E[e^{\alpha\,\int_0^t\,dB_s}\,e^{-\frac12\,\alpha^2t}] = 1
$$

We can move the determinstic term outside of the expectation, and $\int_0^t\,dB_s$ is a normal distribution variable with standard deviation $\sqrt{t}$, we have

$$
e^{\frac12\,\alpha^2\,t} = \E[e^{\alpha\,\sqrt{t}\,Z}] \quad\text{where $Z \sim N(0,1)$}
$$

Q.E.D

Because matrix $\alpha\,B_t$ and $\frac12\,\alpha^2t$ are commutative, this still holds true
$$
\E[e^{\alpha\,B_t - \frac12\,\alpha^2t}] = \E[e^{\alpha\,B_t}\,e^{-\frac12\,\alpha^2t}] = 1
$$

So the equation still holds.

### (a)

If $v_T$ is the underlying asset, $g(v_T)$ is some claim's pay-off based on $v_T$, $f(v,t)$ is the fair value of this claim, we want to write down in this way,

$$
f(v,t)=\E[g(v_T)|\cF_t]
$$

We need $f(v,t)$ to be a martingale, which means if we apply Ito lemma to $f$, we should get $0$ for the drift term. So, we can write down the drite term as,

$$
\Delta = f_t + f_v\,(-\lambda\,(v_t - \bar v)) + \frac12\,(\xi\,v_t)^2\,f_{vv} = 0
$$

So, we have,
$$
\cL = -\lambda\,(v_t - \bar v)\,\partial_v + \frac12\,(\xi\,v_t)^2\,\partial_{vv}
$$

### (b)

We have,
$$
\cL\,f = -\lambda\,(v_t - \bar v)\,\partial_v\,f + \frac12\,(\xi\,v_t)^2\partial_{vv} f
$$

Because $V_0$ and $V_1$ are first order differencial operators, we assume,

\begin{eqnarray}
V_0 = a(v)\,\partial_v \\
V_1 = b(v)\,\partial_v
\end{eqnarray}

So, we have

\begin{eqnarray}
\cL &=& V_0 + \frac12\,V_1^2 \\
&=& a(v)\,\partial_v + \frac12\,b(v)\,\partial_v(b(v)\,\partial_v) \\
&=& a(v)\,\partial_v + \frac12\,b(v)\,\partial_v(b(v))\,\partial_v + \frac12\,b(v)b(v)\,\partial_{vv}
\end{eqnarray}

To match two orders, we have

\begin{eqnarray}
b(v)^2 = (\xi\,v_t)^2 \\
a(v) + \frac12\,b(v)\,\partial_v\,(b(v)) = -\lambda(v_t - \bar v)
\end{eqnarray}

So, we have,
$$
V_0 = (-\lambda(v_t - \bar v) - \frac12\,\xi^2\,v_t)\,\partial_v \\
V_1 = \xi\,v_t\,\partial_v
$$

### (c)

Firstly we try to solve the first ODE, we need to solve,

$$
\frac{d }{dt} v(t) = V_0 v(t) = (-\lambda(v_t - \bar v) - \frac12\,\xi^2\,v_t)\,\partial_v\,(v(t)) = -\lambda(v_t - \bar v) - \frac12\,\xi^2\,v_t
$$

Rearrange it, we have,
$$
\frac{d }{dt} v(t) = -(\lambda + \frac12\,\xi^2)\,v_t + \lambda\,\bar v
$$

which is a linear first order equation, we can use variation of constant to solve it.

Solved the homogeneous equation, we got the general solution,
$$
v(t) = C\,e^{-(\lambda + \frac12\,\xi^2)\,t}
$$

Variation of constant, we assume the particular solution is,
$$
v(t) = C(t)\,e^{-(\lambda + \frac12\,\xi^2)\,t}
$$

Plug into the original ODE, cancel out linear term and rearrange it, we have
$$
C(t)' = \lambda\,\bar v\,e^{(\lambda + \frac12\,\xi^2)t}
$$

Then,
$$
C(t) = \frac{2\lambda\,\bar v}{2\lambda + \xi^2}\,e^{(\lambda + \frac12\,\xi^2)t} + C_1 
$$

Solution is general solution plus the particular solution, we have
$$
v(t) = \frac{2\lambda\,\bar v}{2\lambda + \xi^2} + C_2\,e^{-(\lambda + \frac12\,\xi^2)t}
$$

Plug the boundary condition $v(0) = v_0$, in, we have,
$$
C_2 = v_0 - \bar v' \quad\text{where $\bar v' = \frac{2\lambda\,\bar v}{2\lambda + \xi^2}$}
$$

So, final solution is,
$$
v(t) = \bar v' + (v_0 - \bar v')\,e^{-(\lambda + \frac12\,\xi^2)t} \quad\text{where $\bar v' = \frac{2\lambda\,\bar v}{2\lambda + \xi^2}$}
$$

We have,
$$
e^{V_0\,\Delta}\,v(0) = \bar v' + (v_0 - \bar v')\,e^{-(\lambda + \frac12\,\xi^2)\Delta} \quad\text{where $\bar v' = \frac{2\lambda\,\bar v}{2\lambda + \xi^2}$}
$$


### (c)

Secondly, we are going to solve the second ODE, which is

$$
\frac{d }{dt} v(t) = V_1 v(t) =  \xi\,v_t\,\partial_v\,(v(t)) = \xi\,v_t
$$

This one is easy to solve, use separation of variables, we have,
$$
v(t) = C\,e^{\xi\,t}
$$

Plug into the boundary condition, $v(0) = v_0$, we have,

$$
v(t) = v_0\,e^{\xi\,t}
$$

Evaluated at $t = \sqrt{\Delta}\,Z$, we have,
$$
e^{V_1\sqrt{\Delta}\,Z}\,v(0) = v_0\,e^{\xi\,\sqrt{\Delta}\,Z}
$$

```r
HestonMC2 <- function(params){
    
    is.even <- function(j){as.logical((j+1) %% 2)}
    
    res <- function(T, N, m, evolve,exactVols=NULL)
    {
  
    lambda <<- params$lambda;
    rho <<- params$rho;
    eta <<- params$eta;
    vbar <<- params$vbar;
    v0 <<- params$v;
  
    n <- m*2; #n is number of timesteps = 2*m so we can use Romberg extrapolation
    sqrt2 <- sqrt(2);
    rho2m1 <<- sqrt(1-rho*rho);
    vbarp <<- vbar - eta^2/(4*lambda);
    
    negCount <- 0;
    
    # We use a vertical array, one element per M.C. path
    x <- rep(0,N); v <- rep(1,N)*v0;
    xm <- x; vm <- v;
    W1m <- rep(0,N); W2m <- rep(0,N); 
  
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
        # Two sets of correlated normal random vars.
        
        W1 <- rnorm(N);
        W2 <- rnorm(N);
        W1 <- W1 - mean(W1); W1 <- W1/sd(W1);
        W2 <- W2 - mean(W2); W2 <- W2/sd(W2);
        # Now W1 and W2 are forced to have mean=0 and sd=1
        
        W2p <- W2 - cor(W1,W2)*W1; # Eliminate actual correlation
        W2p <- W2p - mean(W2p); W2 <- W2p/sd(W2p); 
        # Now W1 and W2 have mean=0, sd=1 and correlation=0
        
        L <- rbinom(N, size=1, prob=1/2); # Bernoulli rv for NV step
    
        # Add code for subgrid
        W1m <- W1m + W1/sqrt2; W2m <- W2m + W2/sqrt2; # N(0,1) rv's for subgrid
       
        if (is.even(i)) {
            #print(c(i,mean(W1m),mean(W2m),sd(W1m),sd(W2m),cor(W1m,W2m)));
            resm <- evolve(vm,xm,T/m,W1m,W2m,L);
            xm <- resm[,1];
            vm <- resm[,2];
            W1m <- rep(0,N); W2m <- rep(0,N);
        }
    
        res <- evolve(v,x,T/n,W1,W2,L);
        x <- res[,1];
        v <- res[,2];
        negCount <- negCount +mean(v<0)/n; #Probability of negative variance per path per timestep
    
  }
        
  data.out <- data.frame(x,v,xm) 
  names(data.out) <- c("x","v","xm") 
  return(data.out)         
  
}
return(res)
}
```

```r
# Alfonsi scheme
evolveAlfonsiF <- function(v,x,dt,W1,W2,L=NULL){

        eldt2 <- exp(-lambda*dt/2);

        #Variance process
        vbarp <- vbar - eta^2/(4*lambda);
        psi <- (1-eldt2)/lambda;
        v1 <- v*eldt2+lambda*vbarp*psi;
        v2 <- (v1 > 0) * v1; # Take v2 = 0 if v1<0, else v2=v1
        par <- sqrt(v2) + eta/2 * sqrt(dt)*W2;
        vf <- eldt2*par*par +lambda*vbarp*psi + v1 - v2;
                                # Full truncation

        # Log-stock process (Andersen equation (33))
        vvf <- (v+vf > 0) * (v+vf);
        dw <- vvf/2*dt;
        x <- x  - dw/2 + rho2m1*sqrt(dw)*W1 +
            rho/eta*(lambda*dw + vf-v -lambda*vbar*dt) ;
        # Impose martingale constraint
        x <- x - log(mean(exp(x)));
        v <- vf;
        return(cbind(x,v));
}
```

```r
HestonMC2_old <- function(params){
    
    is.even <- function(j){as.logical((j+1) %% 2)}
    
    res <- function(S0, T, AK, N, m, evolve,exactVols=NULL)
    {
  
    lambda <<- params$lambda;
    rho <<- params$rho;
    eta <<- params$eta;
    vbar <<- params$vbar;
    v0 <<- params$v;
  
    n <- m*2; #n is number of timesteps = 2*m so we can use Romberg extrapolation
    sqrt2 <- sqrt(2);
    rho2m1 <<- sqrt(1-rho*rho);
    vbarp <<- vbar - eta^2/(4*lambda);
    
    negCount <- 0;
    
    # We use a vertical array, one element per M.C. path
    x <- rep(0,N); v <- rep(1,N)*v0;
    xm <- x; vm <- v;
    W1m <- rep(0,N); W2m <- rep(0,N); 
  
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
        # Two sets of correlated normal random vars.
        
        W1 <- rnorm(N);
        W2 <- rnorm(N);
        W1 <- W1 - mean(W1); W1 <- W1/sd(W1);
        W2 <- W2 - mean(W2); W2 <- W2/sd(W2);
        # Now W1 and W2 are forced to have mean=0 and sd=1
        
        W2p <- W2 - cor(W1,W2)*W1; # Eliminate actual correlation
        W2p <- W2p - mean(W2p); W2 <- W2p/sd(W2p); 
        # Now W1 and W2 have mean=0, sd=1 and correlation=0
        
        L <- rbinom(N, size=1, prob=1/2); # Bernoulli rv for NV step
    
        # Add code for subgrid
        W1m <- W1m + W1/sqrt2; W2m <- W2m + W2/sqrt2; # N(0,1) rv's for subgrid
       
        if (is.even(i)) {
            #print(c(i,mean(W1m),mean(W2m),sd(W1m),sd(W2m),cor(W1m,W2m)));
            resm <- evolve(vm,xm,T/m,W1m,W2m,L);
            xm <- resm[,1];
            vm <- resm[,2];
            W1m <- rep(0,N); W2m <- rep(0,N);
        }
    
        res <- evolve(v,x,T/n,W1,W2,L);
        x <- res[,1];
        v <- res[,2];
        negCount <- negCount +mean(v<0)/n; #Probability of negative variance per path per timestep
    
  }

    S <- S0*exp(x);
    Sm <- S0*exp(xm);

    # Now we have three vectors of final stock prices
  
    M <- length(AK);
    AV <- numeric(M); AVdev <- numeric(M);
    BSV <- numeric(M); BSVH <- numeric(M); BSVL <- numeric(M);
    iv2SD <- numeric(M); bias <- numeric(M);
    AVm <- numeric(M); AVmdev <- numeric(M);
    BSVm <- numeric(M); BSVHm <- numeric(M); BSVLm <- numeric(M);
    iv2SDm <- numeric(M);
    AV1 <- numeric(M); AV1dev <- numeric(M);
    BSV1 <- numeric(M); BSVH1 <- numeric(M); BSVL1 <- numeric(M);
    iv2SDrom <- numeric(M);biasRom <- numeric(M);
  
  # Evaluate mean call value for each path
  for (i in 1:M)
  {
    # 2*m timesteps
    K <- AK[i];
    V <- (S>K)*(S - K); # Boundary condition for European call
    AV[i] <- mean(V);
    AVdev[i] <- sqrt(var(V)/length(V));  
    
    BSV[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i]);
    BSVL[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] - AVdev[i]);
    BSVH[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] + AVdev[i]);
    iv2SD[i] <- (BSVH[i]-BSVL[i]);
    
    # m timesteps
    Vm <- (Sm>K)*(Sm - K); # Boundary condition for European call
    AVm[i] <- mean(Vm);
    AVmdev[i] <- sd(Vm) / sqrt(N);
    BSVm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i]);
    BSVLm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] - AVmdev[i]);
    BSVHm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] + AVmdev[i]);
    iv2SDm[i] <- (BSVH[i]-BSVL[i]);
    
    # Richardson extrapolation estimates 
    V1 <- 2*V - Vm;
    AV1[i] <- mean(V1);
    AV1dev[i] <- sd(V1) / sqrt(N);
    BSV1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i]);
    BSVL1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] - AV1dev[i]);
    BSVH1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] + AV1dev[i]);
    iv2SDrom[i] <- (BSVH1[i]-BSVL1[i]);
    
    if(!is.null(exactVols)) {bias <- BSV-exactVols};
    if(!is.null(exactVols)) {biasRom <- BSV1-exactVols};
  }
  
  
  l.AK <- length(AK)      
  data.out <- data.frame(AK,rep(N,l.AK),rep(2*m,l.AK),BSV,bias,iv2SD,BSVm,BSV1,biasRom,iv2SDrom) 
  names(data.out) <- c("Strikes","Paths","Steps","ivol","bias","twoSd","ivolm", "ivolRichardson", "biasRichardson","twoSdRichardson") 
  return(data.out) 
  
  
}
return(res)
}
```

```r
HestonQMC_old <- function(params){
    
    is.even <- function(j){as.logical((j+1) %% 2)}
    
    res <- function(S0, T, AK, N, m, evolve,exactVols=NULL)
    {
  
    lambda <<- params$lambda
    rho <<- params$rho
    eta <<- params$eta
    vbar <<- params$vbar
    v0 <<- params$v
  
    n <- m*2 #n is number of timesteps = 2*m so we can use Romberg extrapolation
    sqrt2 <- sqrt(2)
    rho2m1 <<- sqrt(1-rho*rho)
    vbarp <<- vbar - eta^2/(4*lambda)
    
    negCount <- 0
    
    # We use a vertical array, one element per M.C. path
    x <- rep(0,N); v <- rep(1,N)*v0
    xm <- x; vm <- v
    W1m <- rep(0,N); W2m <- rep(0,N)
    
    # Generation of quasi random numbers now takes place outside the timestep loop
    Z <- sobol(n=N,dim=n,scram=3,seed=4711,norm=T)
      Zperp <- sobol(n=N,dim=n,scram=3,seed=17,norm=T)
      
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
        # Two sets of correlated normal random vars.
        
        W1 <- Z[,i] #Take ith column of pre-generated quasi-rvs
        W2 <- Zperp[,i]
        W1 <- W1 - mean(W1); W1 <- W1/sd(W1)
        W2 <- W2 - mean(W2); W2 <- W2/sd(W2)
        # Now W1 and W2 are forced to have mean=0 and sd=1
        
        W2p <- W2 - cor(W1,W2)*W1 # Eliminate actual correlation
        W2p <- W2p - mean(W2p); W2 <- W2p/sd(W2p) 
        # Now W1 and W2 have mean=0, sd=1 and correlation=0
        
        L <- rbinom(N, size=1, prob=1/2) # Bernoulli rv for NV step
    
        # Add code for subgrid
        W1m <- W1m + W1/sqrt2; W2m <- W2m + W2/sqrt2 # N(0,1) rv's for subgrid
       
        if (is.even(i)) {
            resm <- evolve(vm,xm,T/m,W1m,W2m,L)
            xm <- resm[,1]
            vm <- resm[,2]
            W1m <- rep(0,N); W2m <- rep(0,N);
        }
    
        res <- evolve(v,x,T/n,W1,W2,L)
        x <- res[,1]
        v <- res[,2]
        
  }

    S <- S0*exp(x)
    Sm <- S0*exp(xm)

    # Now we have three vectors of final stock prices
    M <- length(AK);
    AV <- numeric(M); AVdev <- numeric(M);
    BSV <- numeric(M); BSVH <- numeric(M); BSVL <- numeric(M);
    iv2SD <- numeric(M); bias <- numeric(M);
    AVm <- numeric(M); AVmdev <- numeric(M);
    BSVm <- numeric(M); BSVHm <- numeric(M); BSVLm <- numeric(M);
    iv2SDm <- numeric(M);
    AV1 <- numeric(M); AV1dev <- numeric(M);
    BSV1 <- numeric(M); BSVH1 <- numeric(M); BSVL1 <- numeric(M);
    iv2SDrom <- numeric(M);biasRom <- numeric(M);
  
  # Evaluate mean call value for each path
  for (i in 1:M)
  {
    # 2*m timesteps
    K <- AK[i];
    V <- (S>K)*(S - K); # Boundary condition for European call
    AV[i] <- mean(V);
    AVdev[i] <- sqrt(var(V)/length(V));
    BSV[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i]);
    BSVL[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] - AVdev[i]);
    BSVH[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] + AVdev[i]);
    iv2SD[i] <- (BSVH[i]-BSVL[i]);
    
    # m timesteps
    Vm <- (Sm>K)*(Sm - K); # Boundary condition for European call
    AVm[i] <- mean(Vm);
    AVmdev[i] <- sd(Vm) / sqrt(N);
    BSVm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i]);
    BSVLm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] - AVmdev[i]);
    BSVHm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] + AVmdev[i]);
    iv2SDm[i] <- (BSVH[i]-BSVL[i]);
    
    # Richardson extrapolation estimates 
    V1 <- 2*V - Vm
    AV1[i] <- mean(V1)
    AV1dev[i] <- sd(V1) / sqrt(N)
    BSV1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i])
    BSVL1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] - AV1dev[i])
    BSVH1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] + AV1dev[i])
    iv2SDrom[i] <- (BSVH1[i]-BSVL1[i])
    
    if(!is.null(exactVols)) {bias <- BSV-exactVols}
    if(!is.null(exactVols)) {biasRom <- BSV1-exactVols}
  }
  
  data.out <- data.frame(AK,rep(N,M),rep(2*m,M),BSV,bias,iv2SD,BSVm,BSV1,biasRom,iv2SDrom)
  names(data.out) <- c("Strikes","Paths","Steps","ivol","bias","2sd","ivolm", "ivolRichardson", "biasRichardson",
                       "2sdRichardson")
  return(data.out)
  
}
return(res)
}
```
