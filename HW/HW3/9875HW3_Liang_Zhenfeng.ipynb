## Solution

### (a)

Apply Itô's Lemma to to $f = S_t v_t$, we have

\begin{equation}
d(S_t v_t) = S_t\,d v_t + v_t\,d S_t + d [S, v]_t = S_t\,\lambda\,\bar v\,dt - S_t\,\lambda\,v_t\,dt + S_t\,\eta\,\sqrt{v_t}\,dW + V_t\,S_t\,\sqrt{v_t}\,dZ + \rho\,\eta\,S_t\,v_t\,dt\\
\end{equation}


Integrating from 0 to $T$, we have

\begin{equation}
S_T\,v_T - S_0\,v_0 = \lambda\,\bar v\int_0^T S_t\,dt + (\rho\,\eta - \lambda)\int_0^T S_t v_t\,dt + \int_0^T(S_t\,\eta\,\sqrt{v_t}\,dW + V_t\,S_t\,\sqrt{v_t}\,dZ)
\end{equation}

Taking the expection, we have

\begin{eqnarray}
\E[S_T\,v_T] &=& S_0\,v_0 +  \lambda\,\bar v\int_0^T E[S_t]\,dt + (\rho\,\eta - \lambda)\int_0^T E[S_t v_t]\,dt \\
&=& S_0\,v_0 + \lambda\,\bar v\,T\,S_0 + (\rho\,\eta - \lambda)\int_0^T \E[S_t v_t]\,dt \quad\text{($S_t$ is a martingale, so $E[S_t] = S_0$)}
\end{eqnarray}


Let $E[S_t v_t] = \psi (t)$, by taking derivative w.r.t $t$, we have

$$\psi' (t) = \lambda\,\bar v\,S_0 + (\rho\,\eta - \lambda)\psi (t)$$

with $\psi (0) = S_0\,v_0$

Solve the Homogenous Equations to get a General Solution and observed the Particular Solution, we can have the final solution of the ODE is the following,  

$$\psi (T) = E[S_T v_T] = S_0\,v_0\,e^{-(\lambda-\rho\,\eta)\,T}+\lambda\,S_0\,\bar v\,\int_0^T\,e^{-(\lambda-\rho\,\eta)\,(T-t)}\,dt.$$

### (b)

Since $S_t$ is a martingale, so we have
$$
E[S_T] = S_0
$$

From the simplified assumption in the question, we have,
$$
E[v_T] = \bar v
$$

So we have,
\begin{eqnarray}
Cov[S_T, v_T] &=& E[S_T v_T] - E[S_T]E[v_T] \\
&=& S_0\,v_0\,e^{-(\lambda-\rho\,\eta)\,T} + \lambda\,S_0\,\bar v\,\int_0^T\,e^{-(\lambda-\rho\,\eta)\,(T-t)}\,dt - S_0 \bar v \\
&=& S_0\,v_0\,e^{-(\lambda-\rho\,\eta)\,T} + \frac{\lambda\,S_0\,\bar v}{\lambda - \rho\,\eta}(1 - e^{-(\lambda - \rho\,\eta)T}) - S_0\,\bar v \\
&=& S_0\,v_0\,e^{-(\lambda-\rho\,\eta)\,T} + \frac{\lambda\,S_0\,v_0}{\lambda - \rho\,\eta}(1 - e^{-(\lambda - \rho\,\eta)T}) - S_0\,v_0 \quad\text{(Using the assumption $v_0 = \bar v$)} \\
&=& S_0\,v_0 \frac{\rho\,\eta}{\lambda - \rho\,\eta} - S_0\,v_0 \frac{\rho\,\eta}{\lambda - \rho\,\eta} e^{-(\lambda - \rho\,\eta)T} \\
&=& S_0\,v_0 \frac{\rho\,\eta}{\lambda - \rho\,\eta} (1 - e^{-(\lambda - \rho\,\eta)T})
\end{eqnarray}


### (c)

From (b), we have,
$$
Cov[S_T,v_T] = S_0\,v_0 \frac{\rho\,\eta}{\lambda - \rho\,\eta} (1 - e^{-(\lambda - \rho\,\eta)T})
$$

(1) When $T \to 0$, the exponential term will be 1, so, $Cov[S_T,v_T] = 0$ 

(2) When $\rho = 0$, $Cov[S_T,v_T] = 0$ as well. 

(3) $\lambda \to \infty$, exponential term will be zero, and the denominator term will be infinitity, so $Cov[S_T,v_T] = 0$ again. 




### (d)

(1) When $T$ is zero, both $S_T$ and $v_T$ have been determined. Determinate numbers are not correlated.

(2) When both of the underlying driven brownian motion are uncorrelated, for sure these two processes are uncorrelated.

(3) When $\lambda$ is infinitity, it means the mean reverting speed of the variance is infinity which means the $v_t$ might stay at $\bar v$ forever, it will not move at all. So, we basically go back to simplified black scholes model, without drift. $S_t$ is a brownian motion, it is uncorrelated with a constant number, $\bar v$

## Solution

### (a)

```r
impVolSkewHeston <- function(params, T)
{
    
    delta <- 0.001    
    k <- c(-delta, 0., delta)
    
    # Calculate implied vol for the strike vector
    impVols <- impvolHeston(params)(k, T)
    
    # The Central finite approximation to compute the Skew
    impliedSkew <- (impVols[3] - impVols[1]) / (2.0 * delta)
    
    return(impliedSkew)
}

## Call the function
subBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)

## Define a function to calculate the skew for a vector of time to maturity
skew <- function(t){sapply(t,function(x){impVolSkewHeston(subBCC, x)})}
curve(skew(x),from=0.01,to=10,col="blue",ylab="Implied vol. skew",xlab="Time to maturity")
```

### (b)

```r
t <- seq(0.01, 10, length=100)

# log time, in order to get the power
lnx <- log(t)

# log of skew, the original skew calculated is negative, as we can see above, so we need 
# to add a minus sign to make sure it is positive
lny <- log(-1. * skew(t))

# Regression
lm <- lm(lny~lnx) 
lm.sm <- summary(lm)

intercept <- lm.sm$coeff[1]
slope <- lm.sm$coeff[2]

scale <- exp(intercept)
power <- slope

curve(skew(x),from=0.01,to=10,col="red",ylab="Implied vol. skew",xlab="Time to maturity")

# Function of the invert square root of t
empStructure <- function(t){sapply(t,function(x){-1. * scale * x ** (power)})}

# superimpose the square root function onto the ATM skew
curve(empStructure(x),from=0.01,to=10,col="blue",ylab=NA,xlab=NA, add = T)

# add legend to the plot
legend('topleft', c("Heston Skew","1 / sqrt(T)") , lty=1, col=c("red", "blue"), bty='n', cex=.75)
```

#### 3.  (6 points.) 

Code a function in R to return the Heston characteristic function $\phi_T(u)$.  Verify that $\phi_T(0)=1$ and $\phi_T(-i)=1$.  Why must these latter two identities always hold (under the risk neutral measure)?

```r
characFnHeston <- function(params){
    # function to return the Heston characteristic function
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(u,t,j){
        
            al <- -u*u/2 - 1i*u/2 + 1i*j*u;
            bet <- lambda - rho*eta*1i*u-rho*eta*j;
            gam <- eta^2/2;
            d <- sqrt(bet*bet - 4*al*gam);
            rp <- (bet + d)/(2*gam);
            rm <- (bet - d)/(2*gam);
            g <- rm / rp;
            D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
            C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) );
            return(exp(C*vbar + D*v));
    }
    return(res);# Return the Heston characteristic function which takes u, t and j as arguments
}
```

```r
# Initialize the BBC parameters
subBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)

# Charac function phi(u), set t as any number, it doesn't number. j doesn't matter as well
charFn <- function(u){ characFnHeston(subBCC)(u, 0.01, 0) }

cat("We can show in the program, both phi(-i) and phi(0) are 1, the result are", charFn(0), "and", charFn(-1i))
```

By definition,

$$
\phi _T(u):= \EE\left[e^{i u x_T}~|~x_t=0\right]
$$

We have, 

$$
\phi _T(0):= \EE\left[e^{i 0 x_T}~|~x_t=0\right] = \EE\left[1~|~x_t=0\right] = 1
$$

Under risk neutral measure,
$$
\phi _T(-i):= \EE\left[e^{i (-i) x_T}~|~x_t=0\right] = \EE\left[e^{x_T}~|~x_t=0\right] = \frac{1}{K}\EE\left[S_T~|~S_t=K\right] = \frac{1}{K}K = 1
$$

```r
probDensityHeston <- function(params){
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(x,t)
    {
        fj <- characFnHeston(params)

        p <- 1/pi*integrate(function(u){Re(fj(u,t,0)*exp(-1i*u*x))},lower=0,upper=Inf)$value;
        
        return(p);
    }
    
    return(res);# Return Heston density function
}
```

## Solution

### (a)

From Euler scheme, we have

$$
\Delta \hat v_t = \lambda\,(\bar v - v_t)\,\Delta+\eta\,\sqrt{v_t}\,\sqrt{\Delta}\,Z \quad\text{where $Z \sim N(0, 1)$}
$$

$$
\E[\Delta \hat v_t|v_t] = \E[\lambda\,(\bar v - v_t)\,\Delta + \eta\,\sqrt{v_t}\,\sqrt{\Delta}\,Z|v_t] = \lambda\,(\bar v - v_t)\,\Delta
$$

$$
\Var[\Delta \hat v_t|v_t] = \E[(\Delta \hat v_t - \E[\Delta \hat v_t|v_t])^2|v_t] = \E[(\eta\,\sqrt{v_t}\,\sqrt{\Delta}\,Z)^2|v_t] = \E[\eta^2\,v_t\,\Delta\,Z^2|v_t] = \eta^2\,v_t\,\Delta
$$

### (b)

We have,

$$
\E\left[\left.v_{t+\Delta}\right|v_t\right]=\bar v+(v_t-\bar v)\,e^{-\lambda\, \Delta}
$$

$$
\Var[v_{t+\Delta}|v_t] = \frac{\eta ^2}{\lambda}\,  \left(1-e^{- \lambda\,\Delta }\right) \left[ e^{- \lambda\,\Delta }\,v_t+ \frac{1}{2} \left(1-e^{- \lambda\,\Delta }\right)\,\bar v \right]
$$

Plug the exponential term expansion into the above formulas and have some algebra, we have,

$$
\E\left[\left.v_{t+\Delta}\right|v_t\right] 
= v_t + \lambda\,\Delta(\bar v - v_t) + \frac{\lambda^2 \Delta^2}{2}(v_t - \bar v) + \cO(\Delta^3)
$$

\begin{eqnarray}
\Var[v_{t+\Delta}|v_t] = v_t\eta^2\Delta + \frac12\eta^2\lambda\Delta^2(\bar v-3v_t) +  \mathcal{O}(\Delta^3)
\end{eqnarray}

### (c)


$$
\epsilon_E = \E\left[\left.v_{t+\Delta}  - (v_t + \Delta \hat v_t)\right|v_t\right] = \frac{\lambda^2 \Delta^2}{2}(v_t - \bar v)
$$

$$
\epsilon_V = \Var\left[\left.v_{t+\Delta}\right|v_t\right] - \Var\left[(\left. v_t + \Delta \hat v_t)\right|v_t\right] = \frac{\lambda\,\eta^2\,\Delta^2}{2}(\bar v - 3\,v_t)
$$
